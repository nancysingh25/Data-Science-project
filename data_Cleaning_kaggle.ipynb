```python
import pandas as pd
import numpy as np

# read data to pandas
sf_permits = pd.read_csv("C:/Users/HP/Downloads/archive/Building_Permits.csv")
#set random seed
np.random.seed(0)
```

```python
sf_permits.head()
```

```python
# get the number of missing data points per column
missing_values_count = sf_permits.isnull().sum()
#missing_values_count
# look at the # of missing points in the first ten columns
missing_values_count[0:10]
```

```python
# how many total missing values do we have?
total_cells = np.product(sf_permits.shape)
#total_cells
total_missing = missing_values_count.sum()
#total_missing

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

```python
# remove all the rows that contain a missing value
sf_permits.dropna()
```

```python
sf_permits_with_na_dropped = sf_permits.dropna(axis=1)
sf_permits_with_na_dropped.head()
```

```python
#how many columns are dropped
col_original = sf_permits.shape[1]
#col_original
col_dropped_na = sf_permits_with_na_dropped.shape[1]
#col_dropped_na
dropped_columns = col_original- col_dropped_na

dropped_columns
```
